```python
import pandas as pd
```

```python
# Read in January 2019 citibike data and get row count
filename = "201901-citibike-tripdata.csv"
citibike_df = pd.read_csv(filename)
citibike_df.shape[0]
```

```python
# Get a sample of 10% of the dataset
sample_citibike_df = citibike_df.sample(frac=0.1)
sample_citibike_df.shape[0]
```

```python
sample_citibike_df.head()
```

```python
# Convert dataframe to csv
sample_citibike_df.to_csv("tripdata0119.csv", index=False)
```
